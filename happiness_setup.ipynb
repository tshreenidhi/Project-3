```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
```

```python
#Find files
hap_23_path = "Data/World Happiness Report 2023.csv"
hap_22_path = "Data/World Happiness Report 2022.csv"
hap_21_path = "Data/World Happiness Report 2021.csv"
hap_20_path = "Data/World Happiness Report 2020.csv"
hap_19_path = "Data/World Happiness Report 2019.csv"
hap_18_path = "Data/World Happiness Report 2018.csv"
hap_17_path = "Data/World Happiness Report 2017.csv"
hap_16_path = "Data/World Happiness Report 2016.csv"
hap_15_path = "Data/World Happiness Report 2015.csv"
hap_13_path = "Data/World Happiness Report 2013.csv"
sentiment_path = "sentimentdataset.csv"
sent_xref = "sentiment_xref.csv"

#read the csv files
h23_df = pd.read_csv(hap_23_path)
h22_df = pd.read_csv(hap_22_path)
h21_df = pd.read_csv(hap_21_path)
h20_df = pd.read_csv(hap_20_path)
h19_df = pd.read_csv(hap_19_path)
h18_df = pd.read_csv(hap_18_path)
h17_df = pd.read_csv(hap_17_path)
h16_df = pd.read_csv(hap_16_path)
h15_df = pd.read_csv(hap_15_path)
h13_df = pd.read_csv(hap_13_path)
sent_df = pd.read_csv(sentiment_path)
sent_xref = pd.read_csv(sent_xref)
```

```python
# Sort, Rename and add field for 2023 
f_h23_df = h23_df[["Country name", "Ladder score", "Logged GDP per capita", "Generosity", "Social support", "Perceptions of corruption"]]
r_h23_df = f_h23_df.rename(columns={
                    "Country name": "country",
                    "Ladder score": "ladder_score",
                    "Logged GDP per capita": "gdp_per_capita", 
                    "Generosity": "Generosity", 
                    "Social support": "Social support", 
                    "Perceptions of corruption": "corruption"})
r_h23_df.insert(0, "Year", 2023) 
r_h23_df.head()
```
